# GRAFICOS 1

Los graficos son una herramienta muy valiosa, por la cual podemos visualizar la informacion de una manera muy clara y concisa

```python
variable = "rubro_descripcion"
frecuencia = df[variable].value_counts()
plt.figure(figsize=(10, 5))
frecuencia.plot(kind="bar")
plt.title("Frecuencia de los valores únicos en la variable")
plt.xlabel("Valor único")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

por lo que podemos ver, este grafico no es del todo favorable, por lo que devemos utilizar otro tipo de grafico.
Segun nuestro criterio un grafico de torta seria el correcto ya que podemos ver porcentajes y no crear confunciones.

```python
variable = "rubro_descripcion"
frecuencia = df[variable].value_counts()

colors = plt.cm.Set3(range(len(frecuencia)))

plt.figure(figsize=(8, 8))
plt.pie(frecuencia, labels=frecuencia.index, autopct="%1.1f%%", colors=colors)
plt.title("Frecuencia de los rubros 2016 y 2017")

plt.show()
```

En este gráfico podemos visualizar a simple vista cómo están distribuidos los rubros, cual es el que predomina. A su vez también el gráfico nos muestra los porcentajes, esto puede ser muy util a la hora de vizualizar la distribucion y hacerlo con exactitud

```python
variable = "rubro_descripcion"
year_2016 = 2016
year_2017 = 2017
df_2016 = df[df["año"] == year_2016]
df_2017 = df[df["año"] == year_2017]
frecuencia_2016 = df_2016[variable].value_counts()
frecuencia_2017 = df_2017[variable].value_counts()
# Crear los gráficos de torta
plt.figure(figsize=(19, 6))
#año 2016
plt.subplot(1, 2, 1)
plt.pie(frecuencia_2016, labels=None, autopct="%1.1f%%", colors=plt.cm.Set3(range(len(frecuencia_2016))))
plt.title(f"Distribución de rubros {year_2016}", fontsize=16)
#etiquetas
plt.legend(labels=frecuencia_2016.index, bbox_to_anchor=(1, 0.8), loc="center left", fontsize=10)
#2017
plt.subplot(1, 2, 2)
plt.pie(frecuencia_2017, labels=None, autopct="%1.1f%%", colors=plt.cm.Set3(range(len(frecuencia_2017))))
plt.title(f"Distribución de rubros  {year_2017}", fontsize=16)
#etiquetas afuera del gráfico con colores
plt.legend(labels=frecuencia_2017.index, bbox_to_anchor=(1, 0.8), loc="center left", fontsize=10)

plt.tight_layout()
plt.show()
```

En este siguiente grafico dividinos los datos por año y pusimos los rubros en una tabla. De esta manera mantenemos la prolijidad, para no crear confuciones.

#Analizando las zonas economicas

```python
import matplotlib.pyplot as plt

top_10_paises = ['Uruguay', 'Paraguay', 'Chile', 'Bolivia', 'Brasil', 'Perú', 'Estados Unidos', 'Colombia', 'México', 'Ecuador', 'España']
frecuencia_zona_economica = df[df['pais_descripcion'].isin(top_10_paises)]['zona_economica'].value_counts()

plt.figure(figsize=(10, 5))
frecuencia_zona_economica.plot(kind='bar')
plt.title('Frecuencia de Exportaciones por Zona Económica (Top 10 Países)')
plt.xlabel('Zona Económica')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

top_10_paises = ['Uruguay', 'Paraguay', 'Chile', 'Bolivia', 'Brasil', 'Perú', 'Estados Unidos', 'Colombia', 'México', 'Ecuador', 'España']
frecuencia_zona_economica = df[df['pais_descripcion'].isin(top_10_paises)].groupby(['pais_descripcion', 'zona_economica']).size().unstack()

# Ordenar las columnas por frecuencia total en orden descendente
frecuencia_zona_economica = frecuencia_zona_economica[frecuencia_zona_economica.sum().sort_values(ascending=False).index]

fig, ax = plt.subplots(figsize=(10, 5))
frecuencia_zona_economica.plot(kind='bar', ax=ax)
ax.set_title('Frecuencia de Exportaciones por Zona Económica (Top 10 Países)')
ax.set_xlabel('Zona Económica')
ax.set_ylabel('Frecuencia')
ax.legend(title='País')
plt.show()
```
